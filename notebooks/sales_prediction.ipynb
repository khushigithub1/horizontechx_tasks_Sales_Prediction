## Outlier Analysis

Outliers represent unusually high or low observations.

In advertising data, extreme values may correspond to large marketing campaigns rather than errors.

Therefore, outliers should be carefully evaluated before removal.

```python
# Detect Outliers Using IQR
def detect_outliers(column):

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[
        (df[column] < lower) |
        (df[column] > upper)
    ]

    print(f"{column}: {len(outliers)} outliers")


for col in [
    "TV",
    "Radio",
    "Newspaper",
    "Sales"
]:
    detect_outliers(col)
```

```python
# Feature Selection
# Independent Features

X = df[["TV", "Radio", "Newspaper"]]

# Target

y = df["Sales"]

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42
)

print("Training Shape :", X_train.shape)

print("Testing Shape :", X_test.shape)
```

```python
# Train Models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models = {

    "Linear Regression": LinearRegression(),

    "Decision Tree": DecisionTreeRegressor(
        random_state=42
    ),

    "Random Forest": RandomForestRegressor(

        n_estimators=200,

        random_state=42
    )

}
```

```python
# Train All Models
results = []

trained_models = {}

for name, model in models.items():

    model.fit(X_train, y_train)

    prediction = model.predict(X_test)

    mae = mean_absolute_error(y_test, prediction)

    rmse = np.sqrt(mean_squared_error(y_test, prediction))

    r2 = r2_score(y_test, prediction)

    trained_models[name] = model

    results.append({

        "Model": name,

        "MAE": round(mae,3),

        "RMSE": round(rmse,3),

        "R² Score": round(r2,3)

    })

results_df = pd.DataFrame(results)

results_df
```

```python
# Sort Results
results_df = results_df.sort_values(

    by="R² Score",

    ascending=False

)

results_df
```

##  Model Performance Insight

The comparison table evaluates each regression model using three performance metrics:

- MAE

The model with the highest R² score and the lowest prediction errors will be selected as the final prediction model.

```python
# Finding the  Best Model
best_model_name = results_df.iloc[0]["Model"]

best_model = trained_models[best_model_name]

print("="*50)

print("🏆 Best Model")

print("="*50)

print(best_model_name)
```
